This document contains some of the most basic functions used to evaluate values in tables using pandas. A generic CSV file called CSV1 is used as a holder for a comma seperated value text file (.csv)

With pandas you can string together functions using a period (.) and you call attributes using [''] notation. 

```python
import pandas as pd 
```

```python
csv1 = pd.read_csv('CSV1.csv')
```

```python
#To view header information for first two entries the CSV
csv1.head(2)
```

```python
#To see how many entries are available
csv1.info()
```

```python
#To find the average for a column of information 
csv1['Column1'].mean
#To find the maximum value for a column of information 
csv1['Column1'].max
```

```python
#Identify a specific entity
csv1[csv1['Column1'] == 'EntityName']
#Identify another property of the entity
csv1[csv1['Column1'] == 'EntityName']['Column2']
#Identify a entity based on another maximum entity value
csv1[csv1['Column1']==csv1['Column1'].max()]['Entity']
#Alternative method that returns all properties of the entity
csv1.loc[csv1['Column1'].idmax()]
```

```python
#Find average values based on subgroups
csv1.groupby('Subgroup').mean()['Colunn1']
#Number of unique Value
csv1['Column1'].nunique()
```

```python
#Identify number of occurances (Value 2) of subgroup of Value 1
csv1[csv1['Subgroup']==VALUE1]['Column1'].value_counts() == VALUE2
```

```python
#Write a function that returns TRUE of FALSE based on contents of a string
#Lower() -moves everything to lower case
#split() - splits text string
def string_test(Column1):
    if 'content' in Column1.lower().split():
        return True
    else:
        return False
```

```python
#Run function on an example string
string_test('This is an example CONTENT')
#To apply to a Column in the dataset
csv1['Column1'].apply(lambda x: string_test(x))
```
